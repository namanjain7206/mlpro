```python
def convolution(img,img_filter):
    W=img.shape[0]
    H=img.shape[1]
    
    F=img_filter.shape[0]
    new_img=np.zeros((W-F+1,H-F+1))
    for row in range(W-F+1):
        for col in range(F):
            for i in range(F):
                for j in range(F):
                    new_img[row][col]+=img[row+i][col+j]*img_filter[i][j]
                    
                if new_img[row][col]>255:
                    new_img[row][col]=255
                elif new_img[row][col]<0:
                    new_img[row][col]=0
                    
                    
    return new_img
    

    
```

```python
blur_filters=np.ones((3,3))/9.0
output=convolution(img_gray,blur_filters)
plt.imshow(output)
```

```python
edge_filters=np.array(([1,0,-1],[1,0,-1],[1,0,-1]))

output2=convolution(img_gray,edge_filters)
```
